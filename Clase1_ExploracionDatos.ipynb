# CLASE: EXPLORACION DE DATOS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
'Carga de datos de la URL'
url = 'https://raw.githubusercontent.com/DrueStaples/Population_Growth/master/countries.csv'
df_pop = pd.read_csv(url)
df_pop.tail(20)
```

```python
'Busqueda de informacion basica'
df_pop.info() #informacion basica de la data
df_pop.isnull() #para ver si has datos nulos y corregir la data
```

```python
'con este comando nos indica cuantos paises diferentes existe'
paises=pd.get_dummies(df_pop['country'])
paises.info()
```

```python
'Impresion paises encontrados'
columns_names = paises.columns.values
columns_names
```

```python
'ploteo de las poblaciones de todas las ciudades encontradas de la data'
for i in range(141):
  aux=df_pop[df_pop["country"] == columns_names[i] ] #busca los datos de una ciudad particular y los vuelve df
  aux.plot.bar(x='year', y='population')
  plt.title("Poblacion de "+columns_names[i])
```

```python
'Estudio de la poblacion de la ciudad de Peru'
df_pop_pe = df_pop[df_pop["country"] == 'Peru' ]
df_pop_pe
```

```python
'Ploteo de la data de la poblacion de Peru'
df_pop_pe.plot.bar(x='year', y='population')
```

```python
'Estudio de la poblacion de la ciudad de Argentina'
df_pop_ar = df_pop[df_pop["country"] == 'Argentina']
df_pop_ar.head()
```

```python
'Estudio de la poblacion de las ciudad de Argentina-Peru simultaneamente'
anios = df_pop_ar['year'].unique()
pop_ar = df_pop_ar['population'].values
pop_pe = df_pop_pe['population'].values
df_plot = pd.DataFrame({'Argentina': pop_ar,'Peru': pop_pe},index=anios)
df_plot
```

```python
'Ploteo de las poblaciones en conjunto'
df_plot.plot.bar()
plt.title("Poblaciones Peru vs Argentina")
```
