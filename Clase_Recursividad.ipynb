CLASE: ALGORITMO DE ITERACION VS RECURSIVIDAD

```python
def fact_recur(n):
  if n==0:
    res=1
  else:
    res=n*fact_recur(n-1)
  return res
```

```python
def fact_itera(n):
  res=1
  for i in range(1,n+1):
    res=res*i
  return res
```

```python
fact_itera(0)
```

```python
import time
import matplotlib.pyplot as plt

tiempos_iterativos=[]
for i in range(1,100):
  inicio=time.perf_counter()
  fact_itera(i)
  fin=time.perf_counter()
  tiempos_iterativos.append(fin-inicio)


tiempos_recursivos=[]
for i in range(1,100):
  inicio=time.perf_counter()
  fact_recur(i)
  fin=time.perf_counter()
  tiempos_recursivos.append(fin-inicio)
```

```python
plt.plot(tiempos_iterativos)
plt.show()
fact_itera(40)
```

```python
plt.plot(tiempos_recursivos)
plt.show()
fact_recur(40)
```

```python
plt.plot(tiempos_iterativos,label='Iterativo')
plt.plot(tiempos_recursivos,label='Recursivo')
plt.legend()
plt.show()
```

```python
def palin(texto,inicio,fin):
  if inicio>=fin:
    return True
  if texto[inicio]==texto[fin]:
    return palin(texto,inicio+1,fin-1)
  else:
    return False

palin("Hoog",0,3)
```
